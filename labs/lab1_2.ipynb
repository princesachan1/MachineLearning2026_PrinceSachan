```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../dataset/diamonds.csv')

filtered_df = df[(df['clarity'] == 'SI2') & (df['color'] == 'E')]

plt.figure(figsize=(10, 6))
sns.scatterplot(data=filtered_df, x='carat', y='price',hue='cut', s=80)

plt.title('Diamond Price vs Carat', fontsize=14)
plt.xlabel('Carat Size', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.legend(title='Cut Quality')
plt.grid(True, alpha=0.5)
plt.show()
```
