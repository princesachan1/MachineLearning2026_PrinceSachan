```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../dataset/gapminderDataFiveYear.csv')
countries = ['France', 'United Kingdom', 'Italy', 'Germany', 'Spain']
europe_data = df[df['country'].isin(countries)]
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=europe_data, x='year', y='pop', hue='country', marker='o', linewidth=2)

plt.title('Population Trends (1952-2007)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population in Millions', fontsize=12)
plt.xticks(range(1952, 2012, 5))
plt.grid(True, alpha=0.3)
plt.legend(title='Country')
plt.show()
```
