```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="darkgrid")
df = pd.read_csv('../dataset/planets.csv')
mass_data = df[df['mass'].notna()]['mass']
```

```python
plt.figure(figsize=(10, 8))

# Histogram
plt.subplot(2, 1, 1)
sns.histplot(mass_data, kde=False, color='skyblue')
plt.title('Mass Distribution of Planets')
plt.ylabel('Frequency')

# Box Plot
plt.subplot(2, 1, 2)
sns.boxplot(x=mass_data, color='lightcoral')
plt.title('Mass Spread of Planets')
plt.xlabel('Mass (Jupiter Masses)')

plt.tight_layout()
plt.show()
```
